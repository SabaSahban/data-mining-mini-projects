# HW2

## Imports

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

## Load data

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این مرحله همانند تمرین قبلی داده ها را درون یک دیتا فریم pandas ذخیره میکنید
</font>
</p>
</div>

```python
data = pd.read_excel('worldcities.xlsx')
data
```

 ## Add column

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این مرحله با توجه به ستون جمعیت یک ستون جدید تحت عنوان Population Level   اضافه کنید به طوری که به 4 دسته زیر  بر اساس وضعیت هر داده اختصاص داده می شودند.
</br>
    <div dir="ltr">
Low: 0-25%
    </br>
Mid: 25-50%
    </br>
High: 50-70%
    </br>
Over: 70-100%
    </div>
</font>
</p>
</div>

```python
quantiles = data['population'].quantile([0, 0.25, 0.5, 0.7, 1])
bins = [quantiles.iloc[0], quantiles.iloc[1], quantiles.iloc[2], quantiles.iloc[3], quantiles.iloc[4]]
labels = ['Low', 'Mid', 'High', 'Over']
data['Population Level'] = pd.cut(data['population'], bins=bins, labels=labels, include_lowest=True)

data
```

## Preproccess

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این بخش باید ستون های خالی داده ها راحذف کرده و در نهایت تمامی ستون ها را به داده های عددی تبدیل کنید به جز ستون اضافه شده
</font>
</p>
</div>

```python
data = data.dropna()
label_encoder = LabelEncoder()

categorical_columns = ['ville', 'ville_ascii', 'pays', 'iso2', 'iso3', 'admin_nom', 'capital']

for col in categorical_columns:
    data[col] = label_encoder.fit_transform(data[col])
data
```

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 حال باید داده های به دست آمده از قسمت قبلی را نرمال سازی کنید.
</font>
</p>
</div>

```python
columns_to_normalize = [col for col in data.columns if col != 'Population Level']

scaler = MinMaxScaler()
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

data
```

## Split

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
داده ها را به دو بخش تست و آموزش با نسبت 80 به 20 تقسیم کنید.
</font>
</p>
</div>

```python
target = data["population"]
features = data.drop(["population", "Population Level"], axis=1)  
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=50)
```

## Regression

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای این قسمت باید ستون Population  را به عنوان برچسب درنظر گرفته و از بقیه ستون هابه عنوان فیچر استفاده کنید
</font
</p>
</div>

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
باید از رگرسیون خطی
،درجه 1 و درجه 2 استفاده کرده و در هر مورد دقت و خطای میانگین مربعات را بدست آورده و نتایج را باهم مقایسه کنید.</font
</p>
    
</div>

```python
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)

poly_reg_1 = PolynomialFeatures(degree=1)
x_train_poly_1 = poly_reg_1.fit_transform(x_train)
x_test_poly_1 = poly_reg_1.transform(x_test)

linear_reg_poly_1 = LinearRegression()
linear_reg_poly_1.fit(x_train_poly_1, y_train)

poly_reg_2 = PolynomialFeatures(degree=2)
x_train_poly_2 = poly_reg_2.fit_transform(x_train)
x_test_poly_2 = poly_reg_2.transform(x_test)

linear_reg_poly_2 = LinearRegression()
linear_reg_poly_2.fit(x_train_poly_2, y_train)

y_pred_linear = linear_reg.predict(x_test)

y_pred_poly_1 = linear_reg_poly_1.predict(x_test_poly_1)

y_pred_poly_2 = linear_reg_poly_2.predict(x_test_poly_2)

mse_linear = mean_squared_error(y_test, y_pred_linear)

mse_poly_1 = mean_squared_error(y_test, y_pred_poly_1)

mse_poly_2 = mean_squared_error(y_test, y_pred_poly_2)

print(f"Linear Regression MSE: {mse_linear}")
print(f"Polynomial Regression (degree=1) MSE: {mse_poly_1}")
print(f"Polynomial Regression (degree=2) MSE: {mse_poly_2}")
```

## Classification

<div dir="rtl">

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای این قسمت باید ستون Population Level  را به عنوان برچسب درنظر گرفته و از بقیه ستون هابه عنوان فیچر استفاده کنید
همچنین ستون Population را نیز حذف کنید
    </font>

<div dir="rtl">
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">

### Decision Tree (Entropy)

```python
features = data.drop(["population", "Population Level"], axis=1)
target = data["Population Level"]

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=47)

dt_classifier = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_leaf=5, random_state=42)

dt_classifier.fit(x_train, y_train)

y_pred_train = dt_classifier.predict(x_train)
y_pred_test = dt_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of Decision Tree Classifier on Train Set (Entropy): {accuracy_train}")
print(f"Accuracy of Decision Tree Classifier on Test Set (Entropy): {accuracy_test}")
```

### Random Forest (Entropy)

```python
rf_classifier = RandomForestClassifier(n_estimators=1000, max_leaf_nodes=400, n_jobs=-1, criterion='entropy')

rf_classifier.fit(x_train, y_train)

y_pred_train = rf_classifier.predict(x_train)
y_pred_test = rf_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of Random Forest Classifier on Train Set (Entropy): {accuracy_train}")
print(f"Accuracy of Random Forest Classifier on Test Set (Entropy): {accuracy_test}")
```

### KNN K=2

```python
knn_classifier = KNeighborsClassifier(n_neighbors=2)

knn_classifier.fit(x_train, y_train)

y_pred_train = knn_classifier.predict(x_train)
y_pred_test = knn_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of KNN Classifier (K=2) on Train Set: {accuracy_train}")
print(f"Accuracy of KNN Classifier (K=2) on Test Set: {accuracy_test}")
```

### KNN K=3

```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)

knn_classifier.fit(x_train, y_train)

y_pred_train = knn_classifier.predict(x_train)
y_pred_test = knn_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of KNN Classifier (K=3) on Train Set: {accuracy_train}")
print(f"Accuracy of KNN Classifier (K=3) on Test Set: {accuracy_test}")
```

### KNN K=5

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)

knn_classifier.fit(x_train, y_train)

y_pred_train = knn_classifier.predict(x_train)
y_pred_test = knn_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of KNN Classifier (K=5) on Train Set: {accuracy_train}")
print(f"Accuracy of KNN Classifier (K=5) on Test Set: {accuracy_test}")
```

### SVM (Linear)

```python
svm_classifier = SVC(kernel='linear')

svm_classifier.fit(x_train, y_train)

y_pred_train = svm_classifier.predict(x_train)
y_pred_test = svm_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of SVM Classifier (Linear Kernel) on Train Set: {accuracy_train}")
print(f"Accuracy of SVM Classifier (Linear Kernel) on Test Set: {accuracy_test}")
```

### SVM (Non-linear)

```python
svm_classifier = SVC(kernel='rbf')
svm_classifier.fit(x_train, y_train)

y_pred_train = svm_classifier.predict(x_train)
y_pred_test = svm_classifier.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Accuracy of SVM Classifier (RBF Kernel) on Train Set: {accuracy_train}")
print(f"Accuracy of SVM Classifier (RBF Kernel) on Test Set: {accuracy_test}")
```
